```python
from collections import namedtuple
Range = namedtuple('Range', ['start', 'end'])

def overlap(row1, row2):
    r1 = Range(start=row1.start, end=row1.end)
    r2 = Range(start=row2.start, end=row2.end)
    latest_start = max(r1.start, r2.start)
    earliest_end = min(r1.end, r2.end)
    delta = (earliest_end - latest_start).total_seconds()
    overlap = max(0, delta)
    return overlap

def main(df1, df2, col):
    #for cat in df2.consat.unique().tolist():
    for cat in df2[col].unique().tolist():
        df1[cat] = 0
    it1 = df1.iterrows()
    it2 = df2.iterrows()
    idx1, row1 = next(it1)
    idx2, row2 = next(it2)
    while True:
        try:
            r1 = Range(start=row1.start, end=row1.end)
            r2 = Range(start=row2.start, end=row2.end)
            if r2.end < r1.start:
                # no overlap. r2 before r1. advance it2
                idx2, row2 = next(it2)
            elif r1.end < r2.start:
                # no overlap. r1 before r2. advance it1
                idx1, row1 = next(it1)
            else:
                # overlap. overlap(row1, row2) must > 0 
                df1.loc[idx1, row2[col]] += overlap(row1, row2)
                # determine whether to advance it1 or it2
                if r1.end < r2.end:
                    # advance it1
                    idx1, row1 = next(it1)
                else:
                    # advance it2
                    idx2, row2 = next(it2)
        except StopIteration:
            break

```

```python
dict_li=[]

for index, row in consat_df.iloc[:].iterrows():
    start_t=row['start']
    end_t=row['end']
    spireid=row['spire_id']
    contacttype=row['contact_type']
    #gnssid=
    row=row.drop('sum_all')
    idx_sv=row.iloc[4:].gt(0)
    for idx,val in row.iloc[4:][idx_sv].iteritems():
        #print(idx, val)
        gnssid=idx
        
        sample='%s.%03d.%s.%02d.*.%s.SC001_0001' %(start_t.year,start_t.day_of_year,spireid, start_t.hour,gnssid)
        #sample='2023.144.103.01.01.E33.SC001_0001'
        fn='/media/datastore/mirror/spwxdp/repro4/spire/level2/scnLv2/%s/scnLv2_%s.0001_nc'%(sample[:8],sample)
        try:
            lv2=xr.open_mfdataset(fn, decode_times=False)
            #print(index,fn)
            #lv2=load_leo(fn)

            #reset index to actual datetime for slicing
            import datetime
            ro_start = datetime.datetime(year=lv2.year,day=lv2.day,month=lv2.month, hour=lv2.hour, minute=lv2.minute, second=lv2.second)
            lv2['time']=np.datetime64(ro_start) + lv2.time*datetime.timedelta(seconds=1)

            over_ds=lv2.sel(time=slice(start_t, end_t))

            dict_li+=[{"idx":index,"spireid":spireid,
                 "contact_type": contacttype,
                 "time":over_ds.time.values[0],"x":over_ds.x_LEO.values.mean(),
                 "y":over_ds.y_LEO.values.mean(), "z":over_ds.z_LEO.values.mean(),
                 "lat":over_ds.lat.values.mean(), "lon":over_ds.lon.values.mean()}]
        except:
            pass
```

```python
df=pd.DataFrame(dict_li)
```

```python
#Check lat/lon?
import pandas as pd
from os.path import splitext
from pyproj import CRS, Transformer

def xyz_to_latlonalt(x, y, z, crs_in, crs_out):
    crs_from = CRS.from_user_input(crs_in)
    crs_to = CRS.from_user_input(crs_out)
    proj = Transformer.from_crs(crs_from, crs_to, always_xy=True)
    coordinates = proj.transform(x, y, z)
    return coordinates


df['lat1'] = df.apply(lambda row: xyz_to_latlonalt(row['x'], row['y'], row['z'], 25832, 4326)[0], axis=1)
df['lon1'] = df.apply(lambda row: xyz_to_latlonalt(row['x'], row['y'], row['z'], 25832, 4326)[1], axis=1)
df['alt1'] = df.apply(lambda row: xyz_to_latlonalt(row['x'], row['y'], row['z'], 25832, 4326)[2], axis=1)
```
