```python
import xarray as xr
import pandas as pd
from datetime import date, timedelta
import os
```

```python
sdate = date(2022,11,1)   # start date
edate = date(2023,5,1)   # end date
dr=pd.date_range(sdate,edate-timedelta(days=1),freq='d')
dr=["comms_contacts2/"+ i.strftime('%Y-%m-%d') for i in dr]
```

```python
full_list=[]
for direc in dr:
    part_list=os.listdir(direc)
    part_list=[direc + '/' + i for i in part_list]
    full_list+=part_list
```

```python
pd_list=[]
for ncf in full_list:
    ds=xr.open_dataset(ncf)
    #df=ds.to_pandas()
    pd_dict={"start":ds.start, "end":ds.end, "contact_type":ds.contact_type,"spire_id":ds.spire_id}
    df=pd.DataFrame(pd_dict)
    #df["spire_id"]=ds.spire_id
    pd_list+=[df]
```

```python
full_df=pd.concat(pd_list, ignore_index=True)
```

```python
full_df["spire_id"]=pd.to_numeric(full_df.spire_id.str.strip('FM'))
```

```python
full_df.to_pickle("spire_comms_df.pkl")
```

```python
z=full_df.groupby([full_df.spire_id,pd.Grouper(key='start', axis=0, freq='D')]).count().unstack()
```

```python
z=z.end.transpose()
```

```python
z
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots(figsize=(8,4))
ax.fill_between(z.index,z.min(axis=1),z.max(axis=1), color="#1b9e77",alpha=.3,lw=0, label="Range")

ax.plot(z, color="#1b9e77",lw=.3)

ax.plot(z.mean(axis=1), color="#d95f02", lw=5, label="Mean")
ax.set_xlim(z.index[2], z.index[-3])
ax.legend()
ax.set_ylabel("Communications Contacts/ Day")

plt.savefig("../../figures/comms-day.png")
text = fig.text(0.95, 0.02, 'Source: Spire Satellite Operations Metadata comms_contacts_2023_12_12.tar.gz', horizontalalignment='right', wrap=True ) 
fig.tight_layout(rect=(0,.05,1,1)) 
```
